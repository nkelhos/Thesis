```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
from math import pi,exp
```

```python
x = np.linspace(1,20)
y = [ xi**2.2 for xi in x]
fig = plt.figure( figsize=(9,6) )
plt.plot(x,y)
plt.show()
```

```python
def f(s,t):
    oL   = 4
    oD   = 2
    H    = 1.1
    nD   = s[0]
    nL   = s[1]
    dDdt = nL * oL - 3 * H * t
    dLdt = nD * oD - 3 * H * t
    return [dDdt, dLdt]

t  = np.linspace(0.000001,100)
s0 = [20,5]
s  = odeint(f,s0,t)

fig = plt.figure( figsize=(9,6) )
plt.plot( t, s[:,0], 'r--', linewidth=2.0)
plt.plot( t, s[:,1], 'b-' , linewidth=2.0)
#plt.xscale('log')
#plt.yscale('log')
plt.xlabel('t')
plt.ylabel('Number of Particles')
plt.legend(['Dark','SM'])
plt.show()
```

```python
# assume DM is a spin 1/2 Majorana fermion, 
# so X is its own antiparticle and has g=2 degrees of freedom
g = 2 

# sample at several different x's, where x = m/T
x  = np.logspace(-1,4,100)

fig = plt.figure( figsize=(9,6) )

# loop over DM mass m in GeV
for m in [3e0,1e1,3e1,1e2,3e2]:
    
    # when DM is not relativistic (T<~m) , equilibrium number of DM is
    # at temperature T
    def eqT(T):
        return g * (m*T/(2*pi))**(3/2) * exp(-m/T)
    
    eq = [ eqT(m/xi) for xi in x ]
    plt.plot(t, eq, 'g-', label=r'Equilibrium %d GeV'%(m))
    #plt.plot( t, s[:,0], 'r--', linewidth=2.0)

plt.xscale('log')
plt.yscale('log')
plt.xlabel('x=m/T')
plt.ylabel('Number of Particles')
norm = eqT(m/x[0])
plt.xlim([1e-2,1e3])
plt.ylim([norm*1e-20,norm*1e5])
#plt.legend([r'Equilibrium $n_{\chi}$])
plt.title('Equilibrium number density')
plt.legend()
plt.show()
```
